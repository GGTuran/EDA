```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# checking missing values 
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# deleting columns which have null or missing values
df.dropna(axis=1)
```

```python
# Mean value imputation
df["age_mean"] = df['age'].fillna(df["age"].mean())
```

```python
df[['age_mean','age']]
```

```python
# Median value imputation(for outliers) 
df["age_median"] = df["age"].fillna(df["age"].median())
```

```python
df[['age_median','age']]
```

```python
# Mode value imputation(for categorical values)
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df[df['embarked'].notna()]
```

```python
mode_value = df[df['embarked'].notna()]['embarked'].mode()[0]
```

```python
df["embarked_mode"] = df["embarked"].fillna(mode_value)
```

```python
df[['embarked_mode','embarked']]
```
