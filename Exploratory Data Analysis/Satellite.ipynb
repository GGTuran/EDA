```python
# Categorical plot
# Here x is categorical , y is numerical
sns.catplot(x='Class of Orbit', y='Expected Lifetime (yrs.)', data=df, kind="box")
```

```python
# Bar Plot 
plt.bar(df['Class of Orbit'],df['Expected Lifetime (yrs.)'],color='blue')
plt.xlabel("Class of Orbit")
plt.ylabel("Expected Lifetime (yrs.)")
plt.title("Bar Plot")
plt.show()
```

```python
# Histogram 
sns.histplot(df['Expected Lifetime (yrs.)'],bins=10,kde=True)
```

```python
# Kde plot 
sns.kdeplot(df['Launch Mass (kg.)'],fill=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Class of Orbit',y='Expected Lifetime (yrs.)',data=df,estimator=sum)
plt.title("Total Lifetime Per Orbit")
plt.xlabel("Class of Orbit")
plt.ylabel("Expected Lifetime (yrs.)")
plt.show()
```

```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
for col in categorical_features:
    print(df[col].value_counts(normalize=True) * 100)
    print('---------------------------')
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Purpose', 'Class of Orbit']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```

```python
# Finding out the most used Class of Orbit
df['Class of Orbit'].value_counts().plot.pie(y = df['Class of Orbit'], figsize = (10,10), autopct = '%1.1f')
```

```python
df_mass_purpose = df.groupby("Purpose")["Launch Mass (kg.)"].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x="Launch Mass (kg.)", y="Purpose", data=df_mass_purpose)
ax.set_xlabel("Average Launch Mass (kg)")
ax.set_ylabel("")
ax.set_title("Average Launch Mass by Purpose", size=20)
plt.show()
```

```python
df_life_orbit = df.groupby("Class of Orbit")["Expected Lifetime (yrs.)"].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x="Expected Lifetime (yrs.)", y="Class of Orbit", data=df_life_orbit)
ax.set_xlabel("Average Expected Lifetime (Years)")
ax.set_ylabel("")
ax.set_title("Expected Lifetime by Orbit Class", size=20)
plt.show()
```

```python
df_apogee_purpose = df.groupby("Purpose")["Apogee (km)"].sum().sort_values(ascending=False).reset_index().head(10)

plt.figure(figsize=(14, 10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x="Apogee (km)", y="Purpose", data=df_apogee_purpose)
ax.set_xlabel("Total Apogee (km)")
ax.set_ylabel("")
ax.set_title("Top 10 Purposes by Total Apogee", size=20)
plt.show()
```
